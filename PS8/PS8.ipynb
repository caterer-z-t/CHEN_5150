Wittrup Problem 5-2. I expect to see an algebraic equation as the final answer here. 

5-2: Derive an expression for the dynamic protein concentration when transcription is instantaneously and fully inhibited at t = 0. Assume that the system was at steady state prior to inhibition of expression of this particular gene.

mRNA dynamics after inhibition (transcription rate = 0):

Starting from the balance on mRNA in a growing cell volume $V_i$:

$\frac{d([mRNA]V_i)}{dt} = -\gamma_{R}[mRNA]V_i$

$\implies \frac{dV_i}{dt} = \mu V_i$:

$\implies V_i\frac{d[mRNA]}{dt} + [mRNA]\mu V_i = -\gamma_{R}[mRNA]V_i$

$\implies \frac{d[mRNA]}{dt} = -(\gamma_{R} + \mu)[mRNA]$

$\implies [mRNA](t) = [mRNA]_0 \, e^{-(\gamma_{R} + \mu)t}$

Since the system is at steady state at $t = 0$, the initial condition is 
$[mRNA]_0 = [mRNA]_{ss} = \frac{k_R}{\gamma_{R} + \mu}$

$\implies [mRNA](t) = \left(\frac{k_R}{\gamma_{R} + \mu}\right) e^{-(\gamma_{R} + \mu)t}$

Protein dynamics:

$\frac{d[P]}{dt} + (\gamma_{P} + \mu)[P] = k_p[mRNA](t) = \frac{k_P k_R}{\gamma_{R} + \mu} e^{-(\gamma_{R} + \mu)t}$

$\implies \frac{d}{dt}\left([P]e^{(\gamma_P+\mu)t}\right) = \frac{k_P k_R}{\gamma_{R} + \mu} \, e^{(\gamma_P - \gamma_R)t}$

$\implies [P]e^{(\gamma_P+\mu)t} = \frac{k_P k_R}{(\gamma_{R} + \mu)(\gamma_P - \gamma_R)} \, e^{(\gamma_P - \gamma_R)t} + C_1$

$\implies [P](t) = \frac{k_P k_R}{(\gamma_{R} + \mu)(\gamma_P - \gamma_R)} \, e^{-(\gamma_{R}+\mu)t} + C_1 e^{-(\gamma_P+\mu)t}$

$[P](0) = [P]_{ss} = \frac{k_P k_R}{(\gamma_P + \mu)(\gamma_R + \mu)}$:

$\implies \frac{k_P k_R}{(\gamma_P + \mu)(\gamma_R + \mu)} = \frac{k_P k_R}{(\gamma_{R} + \mu)(\gamma_P - \gamma_R)} + C_1$

$C_1 = \frac{k_P k_R}{(\gamma_R + \mu)} \left(\frac{1}{\gamma_P + \mu} - \frac{1}{\gamma_P - \gamma_R}\right) = \frac{-k_P k_R}{(\gamma_P + \mu)(\gamma_P - \gamma_R)}$

$\therefore \boxed{[P](t) = \frac{k_P k_R}{(\gamma_{R} + \mu)(\gamma_{P} - \gamma_{R})} \, e^{-(\gamma_{R} + \mu)t} - \frac{k_P k_R}{(\gamma_{P} + \mu)(\gamma_{P} - \gamma_{R})} \, e^{-(\gamma_{P} + \mu)t}}$

## Problem 2a. 
Derive an expression for the number of cell surface receptors at steady state in the absence of ligand ([equation 5.54](https://read.amazon.com/?asin=B08L9VL33V&_encoding=UTF8&ref=dbs_p_ebk_r00_pbcb_rnvc00)). 

Using equaitons 5.48 - 5.53 at SS all $\partial dx$ eqs = 0. No ligand $\implies [L]_{ss}, [L_e]_{ss}, [C]_{ss}, [C_e]_{ss} = 0$

$\therefore \frac{d[R]}{dt} = v_{s,R} - k_{e,R}[R]_{ss} + k_{rec,R}[R_e]_{ss} = 0$

$\& \therefore \frac{d[R_e]}{dt} = k_{e,R}[R]_{ss} - k_{rec,R}[R_e]_{ss} - k_{deg}[R_e]_{ss} = 0$

$k_{e,R}[R]_{ss} = (k_{rec,R} + k_{deg})[R_e]_{ss} \implies [R_e]_{ss} = \frac{k_{e,R}[R]_{ss}}{k_{rec,R} + k_{deg}}$

$v_{s,R} = k_{e,R}[R]_{ss} - k_{rec,R} \cdot \left(\frac{k_{e,R}[R]_{ss}}{k_{rec,R} + k_{deg}}\right) = k_{e,R}[R]_{ss}\left(\frac{k_{deg}}{k_{rec,R} + k_{deg}}\right)$

$\implies [R]_{ss} = \frac{v_{s,R}(k_{rec,R} + k_{deg})}{k_{e,R}\, k_{deg}} = \boxed{\frac{v_{s,R}}{k_{e,R}}\left( 1 + \frac{k_{rec,R}}{k_{deg}} \right)}$

## Problem 2b
Derive an expression for the fraction of total receptors that reside on the cell surface at steady state in the absence of ligand ([equation 5.55](https://read.amazon.com/?asin=B08L9VL33V&_encoding=UTF8&ref=dbs_p_ebk_r00_pbcb_rnvc00)).

Following information from `problem 2a:`

$[R_e]_{ss} = \frac{k_{e,R}[R]_{ss}}{k_{rec,R} + k_{deg}} \implies \frac{v_{s,R}}{k_{deg}}$

$[R]_{ss} + [R_e]_{ss} = R_{tot} = \frac{v_{s,R}}{k_{e,R}}\left( 1 + \frac{k_{rec,R}}{k_{deg}} \right) + \frac{v_{s,R}}{k_{deg}} = \frac{v_{s,R}}{k_{deg}}\cdot\frac{k_{rec,R} + k_{deg} + k_{e,R}}{k_{e,R}}$

$\frac{[R]_{ss}}{R_{tot}} = \frac{\dfrac{v_{s,R}(k_{rec,R}+k_{deg})}{k_{e,R}\,k_{deg}}}{\dfrac{v_{s,R}(k_{rec,R}+k_{deg}+k_{e,R})}{k_{e,R}\,k_{deg}}}$

$\therefore \boxed{\frac{[R]_{ss}}{R_{tot}} = \frac{k_{rec,R} + k_{deg}}{k_{rec,R} + k_{deg} + k_{e,R}}}$
